# API란 무엇인가?

**API(Application Programming Interface)** 는 서로 다른 소프트웨어 시스템 간에 정보를 주고받거나 기능을 사용할 수 있도록 도와주는 서비스임.

쉽게 말하면, **"요청하면 필요한 데이터를 보내주는 자동화된 시스템"**

## 1. 고양이 사진 API (The Cat API)

### 소개
- 전 세계 고양이 사진을 무료로 제공하는 공개 API입니다.
- 랜덤 고양이 사진이나, 특정 품종의 고양이 사진도 가져올 수 있습니다.

### 공식 문서
- https://docs.thecatapi.com/

### 주요 활용 방법
| 항목 | 내용 |
|:-----|:----|
| URL | `https://api.thecatapi.com/v1/images/search` |
| 요청 방식 | GET |
| 인증키 필요 여부 | 기본 사용은 필요 없음 |
| 반환 데이터 | 고양이 이미지 URL 리스트 (JSON)

### 대표 예제 코드
```python
import requests
from IPython.display import Image, display

url = "https://api.thecatapi.com/v1/images/search"
response = requests.get(url)
data = response.json()
cat_url = data[0]['url']

print("오늘의 고양이 사진:")
display(Image(url=cat_url))
```

### 반환되는 JSON 구조 예시
```json
[
  {
    "id": "MTY3ODIyMQ",
    "url": "https://cdn2.thecatapi.com/images/MTY3ODIyMQ.jpg",
    "width": 1080,
    "height": 1350
  }
]
```

````python
import requests
from IPython.display import Image, display

url = "https://api.thecatapi.com/v1/images/search"
response = requests.get(url)
data = response.json()
cat_url = data[0]['url']

print("오늘의 고양이 사진:")
display(Image(url=cat_url))
````



## 2. 날씨 정보 API (Open-Meteo API)

### 소개
- 위치(위도, 경도)를 기준으로 현재 날씨, 예보, 기온, 풍속 등 다양한 날씨 정보를 제공합니다.
- 무료이며, 인증키 없이 바로 사용할 수 있습니다.

### 공식 문서
- https://open-meteo.com/en/docs

### 주요 활용 방법
| 항목 | 내용 |
|:-----|:----|
| URL | `https://api.open-meteo.com/v1/forecast` |
| 요청 방식 | GET |
| 인증키 필요 여부 | 필요 없음 |
| 반환 데이터 | 날씨 정보 (JSON)

### 대표 예제 코드
```python
import requests
from datetime import datetime

latitude = 37.5665
longitude = 126.9780
today = datetime.utcnow().date()

url = (
    f"https://api.open-meteo.com/v1/forecast?"
    f"latitude={latitude}&longitude={longitude}"
    f"&daily=temperature_2m_max,temperature_2m_min&timezone=Asia%2FSeoul"
)
response = requests.get(url)
data = response.json()

dates = data['daily']['time']
max_temps = data['daily']['temperature_2m_max']
min_temps = data['daily']['temperature_2m_min']

for i, date in enumerate(dates):
    if date == str(today):
        print(f"오늘({today}) 서울 날씨")
        print(f"- 최고 기온: {max_temps[i]}°C")
        print(f"- 최저 기온: {min_temps[i]}°C")
```

### 반환되는 JSON 구조 예시
```json
{
  "daily": {
    "time": ["2025-04-07", "2025-04-08"],
    "temperature_2m_max": [18.5, 20.1],
    "temperature_2m_min": [9.2, 10.5]
  }
}
```


````python
import requests
from datetime import datetime
latitude = 37.5665
longitude = 126.9780
today = datetime.utcnow().date()
#today= '2025-04-14'
url = (
    f"https://api.open-meteo.com/v1/forecast?"
    f"latitude={latitude}&longitude={longitude}"
    f"&daily=temperature_2m_max,temperature_2m_min&timezone=Asia%2FSeoul"
)
response = requests.get(url)
data = response.json()
dates = data['daily']['time']
max_temps = data['daily']['temperature_2m_max']
min_temps = data['daily']['temperature_2m_min']

for date in dates:
    if date == str(today):
        i = dates.index(date)
    
        print(f"오늘({today}) 서울 날씨")
        print(f"- 최고 기온: {max_temps[i]}°C")
        print(f"- 최저 기온: {min_temps[i]}°C")
        break
````

## 3. 명언 제공 API (ZenQuotes API)

### 소개
- 전 세계 다양한 명언을 랜덤하게 제공합니다.
- 별도의 회원가입 없이 사용할 수 있습니다.

### 공식 문서
- https://zenquotes.io/api

### 주요 활용 방법
| 항목 | 내용 |
|:-----|:----|
| URL | `https://zenquotes.io/api/random` |
| 요청 방식 | GET |
| 인증키 필요 여부 | 필요 없음 (하루 50회 제한) |
| 반환 데이터 | 명언 리스트 (JSON)

### 대표 예제 코드
```python
import requests

url = "https://zenquotes.io/api/random"
response = requests.get(url)
data = response.json()

quote = data[0]['q']
author = data[0]['a']

print("오늘의 명언:")
print(f'"{quote}" — {author}')
```

### 반환되는 JSON 구조 예시
```json
[
  {
    "q": "Be yourself; everyone else is already taken.",
    "a": "Oscar Wilde"
  }
]
```

````python
import requests

url = "https://zenquotes.io/api/random"
response = requests.get(url)
data = response.json()
print(data)
import json ## json에 대해서 다양한 기능을 쓸 수 있음
print(json.dumps(data, indent=10)) # dumps를 사용하면 indent를 쓸 수 있음 dumps속성을 사용하면 깔끔하게 정리해줌 indent 들여쓰기 수준이라고 보면댐

quote = data[0]['q']
author = data[0]['a']

print("오늘의 명언:")
print(f'"{quote}" — {author}')
````
